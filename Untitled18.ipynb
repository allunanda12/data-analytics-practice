```python
import statistics
import matplotlib.pyplot as plt
mem_names=['nah','bunne','hone','rith','nehal']
ages=[20,19,19,7,11]
workout=[2.2,2.5,1,1.2,2]
no_of_week_days=[1,2,3,6,5]
calories_burned=[650,300,400,320,220]
print("-----------fitness dataset-------------")
for i in range(len(mem_names)):
  print(mem_names[i],
        ages[i],
        workout[i],
        calories_burned[i])
average_workout=sum(workout)/len(workout)
print(average_workout)
max_calories=max(calories_burned)
print(max_calories)
min_calories=min(calories_burned)
print(min_calories)
median_calories=statistics.median(calories_burned)
print(median_calories)
mstd_dev=statistics.stdev(calories_burned)
print(mstd_dev)
index=calories_burned.index(max_calories)
print("most active member",mem_names[index])
plt.plot(mem_names,calories_burned)
plt.show()
index1=no_of_week_days.index(max(no_of_week_days))
print("high active member",mem_names[index1])
plt.plot(mem_names,no_of_week_days)
plt.show()
```
