```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data={'cust_name':['nan','sam','harry','liik','dfd'],
      'orderid':[1,2,3,4,5],
      'sales':[54,65,87,12,13],
      'price':[264,345,172,541,121],
      'product_name':['watch','cable','shoe','pant','t-shirt'],
      'deliverycharges':[20,15,25,10,30],
      'delaydays':[2,3,1,4,2]}

df=pd.DataFrame(data)
print(df)
```

```python
df1=df.groupby("cust_name")["product_name"].agg(lambda x:x.mode()[0]).reset_index()
print(df1)
```

```python
plt.bar(df['orderid'],df['delaydays'])
```

```python
plt.pie(df['sales'],labels=df['cust_name'])
plt.show()
```

```python
index=df['deliverycharges'].idxmax()
print(df.loc[index,'cust_name'])
```

```python
df2=df.groupby('product_name')['sales'].sum().reset_index()
print(df2)
```

```python
plt.bar(df2['product_name'],df2['sales'])
plt.show()
```

```python
for i in range(len(df['cust_name'])):
  if df['deliverycharges'][i]>15:

    print('deliverycharges',df['deliverycharges'][i],'name of customer',df['cust_name'][i])
    print('\n')
```

```python
 for i in range(len(df['cust_name'])):
  if df['delaydays'][i]>2:
    print('delaydays',df['delaydays'][i],'orderid is',df['orderid'][i])
```
