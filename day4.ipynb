```python
customer_name=['nanda','luca','vicky','don','rithu']
product_purchases=['shoe','shirt','pant','sunscreen','mobile']
cart_abandment=[1,1,0,1,2]
product_buy=[1,2,1,3,0]
customer_ratings=[4,5,2,5,1]
payment_methods=['a','a','b','a','b']
sales=[100,200,150,300,250]
df=pd.DataFrame({'product_purchases':product_purchases,'cart_abandment':cart_abandment,'customer_ratings':customer_ratings,'payment_methods':payment_methods,'product_buy':product_buy,'sales':sales,'customer_name':customer_name})
df
```

```python
d=df.groupby('product_buy')['product_purchases'].count()
print(d)
```

```python
plt.bar(df['product_purchases'],df['product_buy'])
plt.title("quantity of products purchased")
plt.show()
```

```python
plt.pie(df['sales'],labels=df['customer_name'],autopct='%1.1f%%')
plt.title("revenue generated from customer")
plt.show()
```

```python
most_used_payment_method=df['payment_methods'].mode()[0]
print("this is the most used payment method:",most_used_payment_method)
```

```python
plt.plot(df['product_purchases'],df['customer_ratings'])
plt.title("ratings of products")
plt.show()
```
