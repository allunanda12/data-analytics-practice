```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

sales=[12,13,21,45]
profit=[11,12,10,13]
print("median",np.median(profit))
print("mean",np.mean(sales))
print("variance",np.var(sales))
print("std deviation",np.std(profit))
print("corelation",np.corrcoef(profit))
plt.bar(profit,sales)
plt.show()
```
